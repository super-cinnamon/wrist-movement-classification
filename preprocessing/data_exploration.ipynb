# Init

## Imports

```python
import os
import json

import pandas as pd
from skimpy import skim

import seaborn as sns
import matplotlib.pyplot as plt
```

## Variables & Constants

```python
data_dir = "data"
dataset_file_name = "processed_dataset.csv"
moves_file_name = "individual_moves.csv"

DATA_PATH = os.path.join(os.getcwd(), '..', data_dir)

DATASET_FILE = os.path.join(DATA_PATH, dataset_file_name)
MOVES_FILE = os.path.join(DATA_PATH, moves_file_name)
```

```python
dataset = pd.read_csv(DATASET_FILE)
moves = pd.read_csv(MOVES_FILE)
dataset
```

```python
skim(dataset)
```

```python
moves.roll = moves.roll.apply(json.loads)
moves.pitch = moves.pitch.apply(json.loads)
moves.yaw = moves.yaw.apply(json.loads)
moves
```

```python
skim(moves)
```

# Data Exploration

## Individual moves visualization

```python
random_0 = moves[moves["class"] == 0].sample(1)
```

```python
random_1 = moves[moves["class"] == 1].sample(1)
```

```python
# create line plot for each
sns.lineplot(y=random_0.roll.iloc[0], x=range(len(random_0.roll.iloc[0])), errorbar=None)
sns.lineplot(y=random_0.pitch.iloc[0], x=range(len(random_0.pitch.iloc[0])), errorbar=None)
sns.lineplot(y=random_0.yaw.iloc[0], x=range(len(random_0.yaw.iloc[0])), errorbar=None)

plt.legend(["roll", "pitch", "yaw"])
plt.savefig("sample_0.svg", format="svg")
plt.show()
```

```python
# create line plot for each
sns.lineplot(y=random_1.roll.iloc[0], x=range(len(random_1.roll.iloc[0])), errorbar=None)
sns.lineplot(y=random_1.pitch.iloc[0], x=range(len(random_1.pitch.iloc[0])), errorbar=None)
sns.lineplot(y=random_1.yaw.iloc[0], x=range(len(random_1.yaw.iloc[0])), errorbar=None)

plt.legend(["roll", "pitch", "yaw"])
plt.savefig("sample_1.svg", format="svg")
plt.show()
```

## Dataset visualization

### Box plots

```python
# Create a figure with 5 rows and 3 columns
fig, axes = plt.subplots(nrows=9, ncols=3, figsize=(15, 10))

# Iterate over the columns, skipping the last one
for i, col in enumerate(dataset.columns[:-1]):
    row = i // 3
    col_idx = i % 3
    sns.boxplot(x=dataset[col], ax=axes[row, col_idx])
    axes[row, col_idx].set_title(col)

# Adjust spacing between subplots
plt.tight_layout()
plt.savefig("box_plots.svg", format="svg")
plt.show()
```

### Distribution

```python
dataset["class"].value_counts()
```

```python
# class distribution
sns.histplot(dataset["class"], bins=3)
plt.xticks([0, 1])
plt.savefig("distribution.svg", format="svg")
plt.show()
```

### Correlation

```python
correlation_matrix = dataset.corr()

sns.heatmap(correlation_matrix)
plt.savefig("correlation.svg", format="svg", bbox_inches='tight')
plt.show()
```
